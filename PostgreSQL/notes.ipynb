# Start

```python
from psycopg2 import connect
```

```python
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
conn = connect(
    host="localhost",
    port=5433,
    user="postgres",
    password=os.getenv("postgres_pw"),
    database="test"
)
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("""
    CREATE TABLE cpus (
        id SERIAL PRIMARY KEY,
        cores int,
        name VARCHAR(255),
        threads int,
        prices decimal
    );
""")
```

```python
conn.commit()
```

```python
conn.close()
```

```python
def get_connection():
    return connect(
        host="localhost",
        port=5433,
        user="postgres",
        password=os.getenv("postgres_pw"),
        database="test"
    )
```

```python
with get_connection() as conn:
    with conn.cursor() as cur:
        cur.execute("""
            CREATE TABLE motherboards(
                id SERIAL PRIMARY KEY,
                name VARCHAR(255),
                form_factor VARCHAR(255),
                price decimal
            );
        """)
```

```python
# Commit in psycopg2 with context connection is automatic.
```

# Parameterization


```python
some_cpus = [
    ("Intel Core i9-10900K", 10, 20, 499.99),
    ("AMD Ryzen 9 5950X", 16, 32, 799.99),
    ("Intel Core i7-10700K", 8, 16, 359.99),
    ("AMD Ryzen 7 5800X", 8, 16, 449.99)
]

some_motherboards = [
    {
        "name": "ASUS ROG Strix Z590-E Gaming",
        "form_factor": "ATX",
        "price": 279.99
    },
    {
        "name": "GIGABYTE Z590 AORUS MASTER",
        "form_factor": "EATX",
        "price": 499.99
    },
    {
        "name": "MSI MPG Z590 GAMING EDGE WIFI",
        "form_factor": "ATX",
        "price": 249.99
    }
]
```

```python
with get_connection() as conn:
    with conn.cursor() as cur:
        cur.executemany("""
            INSERT INTO cpus (name, cores, threads, prices)
            VALUES (%s, %s, %s, %s);
        """, some_cpus)
```

```python
with get_connection() as conn:
    with conn.cursor() as cur:
        cur.executemany("""
            INSERT INTO motherboards (name, form_factor, price)
            VALUES (%(name)s, %(form_factor)s, %(price)s);
        """, some_motherboards)
```
