```python
#safer approach 
#positional placeholder -> ?
flavor = input("Enter a flavor:")
rating = input("Enter a rating:")

cur.execute(f"INSERT INTO ice_cream_flavors (flavor,rating) VALUES (?, ?);" , (flavor,rating))

conn.commit()

#sqlite3

```

```python
#named placeholders -> :name

flavor = input("Enter a flavor:")
rating = input("Enter a rating:")

cur.execute(f"INSERT INTO ice_cream_flavors (flavor,rating) VALUES (:flavor, :rating);" , 
                {
                    "flavor":flavor,
                    "rating": rating
                })

conn.commit()
```

```python
# executemany

ice_cream_flavors = [
    ('vanilla', 7),
    ('mint', 2),
    ('caramel',7),
]

cur.executemany(f"INSERT INTO ice_cream_flavors (flavor,rating) VALUES (?, ?);" , ice_cream_flavors)

conn.commit()
```
